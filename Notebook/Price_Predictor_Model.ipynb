## Data Cleaning

- Year Column

```python
dataset = dataset[dataset['year'].str.isnumeric()]
```

```python
dataset['year'] = dataset['year'].astype(int)
```

```python
dataset.info()
```

- Price

```python
dataset = dataset[dataset['Price'] != 'Ask For Price']
```

```python
dataset['Price'] = dataset['Price'].str.replace(',','').astype(int)
```

```python
dataset.info()
```

- KMS Driven

```python
dataset['kms_driven'] = dataset['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')
```

```python
dataset = dataset[dataset['kms_driven'].str.isnumeric()]
```

```python
dataset['kms_driven'] = dataset['kms_driven'].astype(int)
```

```python
dataset.info()
```

* Fuel Type

```python
dataset = dataset[~dataset['fuel_type'].isna()]    # ~ helps in Excluding the Row
```

```python
dataset.isnull().sum()
```

```python
dataset.isna().sum()
```

* Name

```python
dataset['name'] = dataset['name'].str.split(' ').str.slice(0, 3).str.join(' ')
```

```python
dataset.head()
```

* Reset Index

```python
# dataset.reset_index()
```

```python
dataset = dataset.reset_index(drop=True)
```

```python
dataset
```

```python
dataset.describe()
```

* Removing Outliers

```python
dataset[dataset['Price'] > 5e6]
```

```python
dataset = dataset[dataset['Price'] < 5e6].reset_index(drop = True)
```

## Model Building

```python
x = dataset.drop(columns = 'Price')
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(x[['name', 'company', 'fuel_type']])
```

```python
ohe.categories_
```

```python
column_transform = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name', 'company', 'fuel_type']), remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipeline = make_pipeline(column_transform, lr)
```

```python
pipeline.fit(x_train, y_train)
```

```python
pipeline.score(x_test, y_test) * 100
```

```python
y_pred = pipeline.predict(x_test)
```

```python
r2_score(y_test, y_pred) * 100
```

```python
scores = [] 
for i in range(1000):
    X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=i)
    Lrg = LinearRegression()
    pipe = make_pipeline(column_transform, Lrg) 
    pipe.fit(X_train, Y_train) 
    scores.append(pipe.score(X_test, Y_test) * 100)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=np.argmax(scores))
Lrg = LinearRegression()
pipe = make_pipeline(column_transform, Lrg) 
pipe.fit(X_train, Y_train) 
Y_pred = pipe.predict(X_test) 
r2_score(Y_test, Y_pred) * 100
```
